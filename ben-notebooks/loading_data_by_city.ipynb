```python
import json
import pandas as pd
```

```python
# Raw reviews data file as received from Yelp
all_reviews_file = '../raw-data/yelp_academic_dataset_review.json'
all_businesses_file = '../raw-data/yelp_academic_dataset_business.json'
```

```python
def get_df(json_file_name, max_rows=None, city=None, select_keys=None):
    with open(json_file_name, 'r') as f:
        i_row = 0
        df_dict_list = []
        for line in f:
            row_dict = json.loads(line)
            row_city = row_dict.get('city', '')
            if select_keys is not None:
                row_dict = {k: row_dict[k] for k in select_keys}
            if (city is None) or (city == row_city):
                df_dict_list.append(row_dict)
                i_row += 1
            if (max_rows is not None) and (i_row >= max_rows):
                break
        df = pd.DataFrame(df_dict_list)
        return df
```

# Measure read-in time for different amounts of lines read in

```python
n_vals = [10000, 50000, 100000]
for n in n_vals:
    print('n = {}'.format(n))
    %timeit -r1 get_df(all_businesses_file, n)
print('n = all')
%timeit -r1 get_df(all_businesses_file)

# n = 10000
# 1 loop, best of 1: 284 ms per loop
# n = 50000
# 1 loop, best of 1: 1.5 s per loop
# n = 100000
# 1 loop, best of 1: 2.97 s per loop
# n = all
# 1 loop, best of 1: 4.15 s per loop
```

```python
df_full = get_df(all_businesses_file)
print('{} total businesses'.format(len(df_full)))
df_full.info(memory_usage='deep')
del df_full
```

```python
n_vals = [10000, 50000, 100000]
for n in n_vals:
    print('n = {}'.format(n))
    %timeit -r1 get_df(all_businesses_file, n, select_keys=['name', 'city', 'state'])
print('n = all')
%timeit -r1 get_df(all_businesses_file, select_keys=['name', 'city', 'state'])

# n = 10000
# 1 loop, best of 1: 233 ms per loop
# n = 50000
# 1 loop, best of 1: 1.23 s per loop
# n = 100000
# 1 loop, best of 1: 2.5 s per loop
# n = all
# 1 loop, best of 1: 3.46 s per loop
```

Reading in the data is actually faster if we select a subset of columns to keep rather than load the whole thing, then pick a subset afterwards. No problems loading the businesses data into Jupyter either way.
